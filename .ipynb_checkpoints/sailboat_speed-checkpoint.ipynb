```python
import pandas as pd
import numpy as np
from scipy.interpolate import RegularGridInterpolator

# =========================================================
# CONFIG
# =========================================================
KNOT_TO_MS = 0.51444
DEBUG = False

# =========================================================
# HELPERS
# =========================================================
def to_numeric_array(arr):
    return pd.to_numeric(arr, errors="coerce").to_numpy(dtype=float)

# =========================================================
# DATA LOADING
# =========================================================
def load_polar_data(polar_csv, rb_csv):
    # --- LOAD MAIN POLAR CSV ---
    df = pd.read_csv(polar_csv, header=None, dtype=str, skip_blank_lines=False)

    # True Wind Speeds
    TWS_knots = to_numeric_array(df.iloc[0, 2:])
    TWS = TWS_knots * KNOT_TO_MS

    # True Wind Angles
    TWA = to_numeric_array(df.iloc[4:15, 1])

    # Speed table (m/s)
    Vs_table = (
        df.iloc[4:15, 2:2 + len(TWS)]
        .apply(pd.to_numeric, errors="coerce")
        .to_numpy(dtype=float) * KNOT_TO_MS
    )

    # Sail areas
    upwind_sail_area = pd.to_numeric(df.iloc[23, 5], errors="coerce")
    downwind_sail_area = pd.to_numeric(df.iloc[24, 5], errors="coerce")

    # Ensure contiguous arrays
    TWS = np.ascontiguousarray(TWS)
    TWA = np.ascontiguousarray(TWA)
    Vs_table = np.ascontiguousarray(Vs_table)

    # Interpolator
    interp_polar = RegularGridInterpolator(
        (TWA, TWS),
        Vs_table,
        bounds_error=False,
        fill_value=None
    )

    # --- LOAD BEATING / RUNNING CSV ---
    df_rb = pd.read_csv(rb_csv, header=None, dtype=str, skip_blank_lines=False)
    df_rb_num = df_rb.iloc[:, 1:].apply(pd.to_numeric, errors="coerce")

    TWS_rb_knots = to_numeric_array(df_rb.iloc[1, 2:])
    TWS_rb = TWS_rb_knots * KNOT_TO_MS

    beating_TWA, beating_Vs = [], []
    running_TWA, running_Vs = [], []

    for i, row in enumerate(df_rb.itertuples(index=False)):
        label = str(row[0])
        TWA_val = df_rb_num.iloc[i, 0]

        if np.isnan(TWA_val):
            continue

        values = df_rb_num.iloc[i, 1:].to_numpy(dtype=float)
        valid_idx = np.where(~np.isnan(values))[0]

        if len(valid_idx) == 0:
            continue

        j = valid_idx[0]
        tws = TWS_rb[j]
        Vs_val = values[j] * KNOT_TO_MS

        if "Beating" in label:
            beating_TWA.append((tws, TWA_val))
            beating_Vs.append((tws, Vs_val))

        elif "Running" in label:
            running_TWA.append((tws, TWA_val))
            running_Vs.append((tws, Vs_val))

    beating_TWA = np.array(beating_TWA)
    beating_Vs = np.array(beating_Vs)
    running_TWA = np.array(running_TWA)
    running_Vs = np.array(running_Vs)

    if DEBUG:
        print("Loaded polar data")
        print("TWS range:", TWS[0], "-", TWS[-1])
        print("TWA range:", TWA[0], "-", TWA[-1])

    return {
        "TWS": TWS,
        "TWA": TWA,
        "interp_polar": interp_polar,
        "beating_TWA": beating_TWA,
        "beating_Vs": beating_Vs,
        "running_TWA": running_TWA,
        "running_Vs": running_Vs,
        "upwind_sail_area": upwind_sail_area,
        "downwind_sail_area": downwind_sail_area
    }


# =========================================================
# SPEED COMPUTATION
# =========================================================
def compute_Vs(TWA_input, TWS_input, data):
    TWS = data["TWS"]
    TWA = data["TWA"]
    interp_polar = data["interp_polar"]

    beating_TWA = data["beating_TWA"]
    beating_Vs = data["beating_Vs"]
    running_TWA = data["running_TWA"]
    running_Vs = data["running_Vs"]

    # --- HARD TWS CUT ---
    if TWS_input < TWS[0] or TWS_input > TWS[-1]:
        return 0.0

    # --- INTERPOLATE LIMITS ---
    TWA_beat = np.interp(TWS_input, beating_TWA[:, 0], beating_TWA[:, 1])
    TWA_run = np.interp(TWS_input, running_TWA[:, 0], running_TWA[:, 1])

    TWA_min = TWA[0]
    TWA_max = TWA[-1]

    # --- HARD TWA CUTS ---
    if TWA_input < TWA_beat or TWA_input > TWA_run:
        return 0.0

    # --- MAIN INTERPOLATION ---
    point = (TWA_input, TWS_input)
    Vs_main = interp_polar(point)

    # Precompute boundary values
    Vs_min = interp_polar((TWA_min, TWS_input))
    Vs_max = interp_polar((TWA_max, TWS_input))

    # --- UPWIND REGION ---
    if TWA_input < TWA_min:
        Vs_beat = np.interp(TWS_input, beating_Vs[:, 0], beating_Vs[:, 1])

        Vs = Vs_beat + (Vs_min - Vs_beat) * (
            (TWA_input - TWA_beat) / (TWA_min - TWA_beat)
        )

    # --- DOWNWIND REGION ---
    elif TWA_input > TWA_max:
        Vs_run = np.interp(TWS_input, running_Vs[:, 0], running_Vs[:, 1])

        Vs = Vs_max + (Vs_run - Vs_max) * (
            (TWA_input - TWA_max) / (TWA_run - TWA_max)
        )

    # --- INSIDE TABLE ---
    else:
        Vs = Vs_main

    if DEBUG:
        print(f"TWA={TWA_input}, TWS={TWS_input} → Vs={Vs:.3f} m/s")

    return float(Vs)


# =========================================================
# EXAMPLE USAGE
# =========================================================
if __name__ == "__main__":
    data = load_polar_data(
        "inputs/polars_hallberg_rassy_342.csv",
        "inputs/beating_running.csv"
    )

    # Example input (m/s)
    TWA_input = 140
    TWS_input = 4 * KNOT_TO_MS  # convert if needed

    Vs = compute_Vs(TWA_input, TWS_input, data)
    Vs_m_s = float(Vs)
    Vs_knots = Vs_m_s / KNOT_TO_MS

    print(f"\nSailboat speed: {Vs_m_s:.3f} m/s ({Vs_knots:.3f} knots)")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data first
data = load_polar_data(
    "inputs/polars_hallberg_rassy_342.csv",
    "inputs/beating_running.csv"
)

# Define ranges
TWA_range = np.arange(35, 176, 1)
TWS_range_knots = np.arange(4, 26, 1)

plt.figure(figsize=(12, 7))

for tws_knots in TWS_range_knots:
    TWS_ms = tws_knots * KNOT_TO_MS

    Vs_line = [
        compute_Vs(twa, TWS_ms, data) / KNOT_TO_MS  # convert to knots for plotting
        for twa in TWA_range
    ]

    plt.plot(TWA_range, Vs_line, label=f"{tws_knots} kt")

plt.xlabel("True Wind Angle (deg)")
plt.ylabel("Sailboat Speed (knots)")
plt.title("Polar Curves (Interpolated)")

plt.legend(title = "True Wind Speed", ncol=2, fontsize=8)
plt.grid(True)
plt.show()
```

```python
import math
import numpy as np

def relative_angle(direction, heading):
    """Compute relative angle between wind/wave and boat heading in degrees (0-180)."""
    angle = abs(direction - heading) % 360
    return angle if angle <= 180 else 360 - angle


def sailboat_speed_polar_physics(
    wind_speed, wind_dir, boat_heading,
    wave_height, wave_dir,
    LOA, beam, draft, displaced_weight,
    data,
    rho_water=1005.0, nu_water=1e-6, k_w=0.006
):
    """
    Returns sailboat speed (m/s) using compute_Vs polar speed with physics-based
    wave resistance and following-sea surfing boost applied.

    Parameters
    ----------
    wind_speed : float
        True wind speed (m/s)
    wind_dir : float
        True wind direction (deg)
    boat_heading : float
        Boat heading (deg)
    wave_height : float
        Significant wave height (m)
    wave_dir : float
        Wave propagation direction (deg)
    LOA, beam, draft : float
        Hull dimensions (m)
    displaced_weight : float
        Boat displacement (kg)
    data : dict
        Polar data loaded via load_polar_data
    rho_water : float
        Water density (kg/m³)
    nu_water : float
        Kinematic viscosity of water (m²/s)
    k_w : float
        Empirical wave-making resistance coefficient
    """
    # --- Relative angles ---
    twa = relative_angle(wind_dir, boat_heading)
    wave_angle = relative_angle(wave_dir, boat_heading)

    # --- Compute polar speed using compute_Vs ---
    Vs = compute_Vs(twa, wind_speed, data)

    # --- Hull coefficients ---
    displaced_volume = displaced_weight / rho_water
    Cb = displaced_volume / (LOA * beam * draft)
    Cb = np.clip(Cb, 0.1, 1.0)

    # --- Physics-based wave resistance ---
    S_wetted = 2 * (LOA * draft + draft * beam)  # rough wetted area
    Re = Vs * LOA / nu_water
    Cf = 0.075 / (math.log10(Re) - 2)**2 if Re > 0 else 0.0
    R_friction = 0.5 * rho_water * Vs**2 * S_wetted * Cf
    R_wave_making = k_w * rho_water * 9.81 * Vs**2 * LOA**2

    # Added wave resistance
    c = math.sqrt(9.81 * LOA / (2 * math.pi))  # approximate wave phase speed
    theta_w_rad = math.radians(wave_angle)
    if wave_angle < 90:  # head seas
        f_wave = 1 + (Vs / c) * math.cos(theta_w_rad)
    else:  # following seas
        f_wave = max(0.0, 1 - 0.5 * (Vs / c) * math.cos(theta_w_rad))
    R_wave = 0.5 * rho_water * 9.81 * wave_height**2 * LOA * f_wave

    # --- Total resistance ---
    R_total = R_friction + R_wave_making + R_wave
    if R_total > 0:
        Vs_corrected = Vs * (Vs / (Vs + R_total / (0.5 * rho_water * Vs)))  # modest adjustment
    else:
        Vs_corrected = Vs

    # --- Surfing boost for following seas ---
    if 90 < wave_angle <= 180:
        ks = 0.5
        Vs_corrected += ks * max(0, c - Vs_corrected) * 0.2  # empirical boost

    # --- Limit to hull speed ---
    hull_speed = 1.34 * math.sqrt(LOA) * 0.51444
    Vs_corrected = min(Vs_corrected, hull_speed * 1.3)

    return Vs_corrected
```

```python
## Wind and wave affect on sailboat speed (not using polars)
import math
import numpy as np

def relative_angle(direction, heading):
    """Compute relative angle between wind/wave and boat heading in degrees (0-180)."""
    angle = abs(direction - heading) % 360
    return angle if angle <= 180 else 360 - angle

def sailboat_speed_full_physics(
    wind_speed, wind_dir, boat_heading,
    wave_height, wave_dir,
    LOA, beam, draft, displaced_weight,
    sail_area_upwind, sail_area_downwind,
    C_L=1.0, rho_air=1.225, rho_water=1005.0,
    nu_water=1e-6, k_w=0.006, wave_period=None
):
    """
    Estimate sailboat speed (m/s) including:
    - Sail driving force from apparent wind
    - Hull resistance (friction + wave-making)
    - Added wave resistance (head/following seas)
    - Wave surfing boost for following seas
    Iteratively solves for Vb where Fsail = Rtotal
    """
    # --- Relative angles ---
    twa = relative_angle(wind_dir, boat_heading)
    wave_angle = relative_angle(wave_dir, boat_heading)
    
    # --- Select sail area ---
    if twa < 90:  # upwind
        A_sail = sail_area_upwind
    else:
        A_sail = sail_area_downwind
    
    # --- Initial guess for boat speed ---
    Vb = 1.0  # m/s
    for _ in range(20):  # iterative solver
        # --- Apparent wind ---
        Va = math.sqrt(wind_speed**2 + Vb**2 - 2*wind_speed*Vb*math.cos(math.radians(twa)))
        # --- Sail driving force ---
        Fsail = 0.5 * rho_air * C_L * A_sail * Va**2
        
        # --- Hull resistance ---
        S_wetted = 2 * (LOA * draft + draft * beam)  # rough wetted surface area m²
        Re = Vb * LOA / nu_water
        Cf = 0.075 / (math.log10(Re) - 2)**2 if Re>0 else 0.0
        R_friction = 0.5 * rho_water * Vb**2 * S_wetted * Cf
        R_wave_making = k_w * rho_water * 9.81 * Vb**2 * LOA**2  # simple wave-making
        
        # --- Added resistance from waves ---
        if wave_period is not None:
            # Compute wave phase speed
            k = 2 * math.pi / LOA  # approximate wavelength = LOA
            c = math.sqrt(9.81 * wave_period / (2*math.pi) * math.tanh(k * draft))
        else:
            c = math.sqrt(9.81 * LOA / (2*math.pi))  # rough approximation
        # Empirical factor f(Vb/c, θw)
        theta_w_rad = math.radians(wave_angle)
        if wave_angle < 90:
            f_wave = 1 + (Vb/c)*math.cos(theta_w_rad)  # head seas increases resistance
        else:
            f_wave = max(0.0, 1 - 0.5*(Vb/c)*math.cos(theta_w_rad))  # following seas reduces resistance
        R_wave = 0.5 * rho_water * 9.81 * wave_height**2 * LOA * f_wave
        
        # --- Total resistance ---
        R_total = R_friction + R_wave_making + R_wave
        
        # --- Wave surfing boost for following seas ---
        if 90 < wave_angle <= 180:  # mostly following
            ks = 0.5
            Vb += ks * max(0, c - Vb) * 0.2  # modest boost
        
        # --- Update boat speed ---
        Vb_new = Vb * (Fsail / R_total) if R_total>0 else 0.0
        # --- Convergence check ---
        if abs(Vb_new - Vb) < 0.01:
            break
        Vb = Vb_new
    
    # --- Limit to hull speed ---
    hull_speed = 1.34 * math.sqrt(LOA) * 0.51444  # m/s
    Vb = min(Vb, hull_speed * 1.3)  # allow small surfing above hull speed
    
    return Vb
```

```python
# =========================================================
# EXAMPLE USAGE
# =========================================================
if __name__ == "__main__":
    # -----------------------------
    # Load polar data
    # -----------------------------
    data = load_polar_data(
        "inputs/polars_hallberg_rassy_342.csv",
        "inputs/beating_running.csv"
    )

    # -----------------------------
    # Example inputs
    # -----------------------------
    TWA_input = 90                  # True Wind Angle [deg]
    TWS_input_knots = 20              # True Wind Speed [knots]
    TWS_input = TWS_input_knots * KNOT_TO_MS  # convert to m/s

    wave_height = 0.5                # m
    wave_dir = 180                    # deg
    boat_heading = 0                # deg

    # Boat parameters
    LOA = 10.5      # m
    beam = 3.5      # m
    draft = 2.0     # m
    displacement = 7000  # kg
    sail_area_upwind = 50  # m²
    sail_area_downwind = 80  # m²

    # -----------------------------
    # 1. Compute polar speed (m/s)
    # -----------------------------
    Vs_polar = compute_Vs(TWA_input, TWS_input, data)
    print(f"Polar sailboat speed: {Vs_polar:.3f} m/s "
          f"({Vs_polar / KNOT_TO_MS:.3f} knots)")

    # -----------------------------
    # 2. Compute polar + physics corrected speed
    # -----------------------------
    Vs_polar_physics = sailboat_speed_polar_physics(
        wind_speed=TWS_input,
        wind_dir=TWS_input,  # assuming wind from same direction as input
        boat_heading=boat_heading,
        wave_height=wave_height,
        wave_dir=wave_dir,
        LOA=LOA,
        beam=beam,
        draft=draft,
        displaced_weight=displacement,
        data=data
    )
    print(f"Polar + physics sailboat speed: {Vs_polar_physics:.3f} m/s "
          f"({Vs_polar_physics / KNOT_TO_MS:.3f} knots)")

    # -----------------------------
    # 3. Full physics model (no polars)
    # -----------------------------
    Vs_full_physics = sailboat_speed_full_physics(
        wind_speed=TWS_input,
        wind_dir=TWS_input,  # assuming wind from same direction as input
        boat_heading=boat_heading,
        wave_height=wave_height,
        wave_dir=wave_dir,
        LOA=LOA,
        beam=beam,
        draft=draft,
        displaced_weight=displacement,
        sail_area_upwind=sail_area_upwind,
        sail_area_downwind=sail_area_downwind
    )
    print(f"Full physics sailboat speed: {Vs_full_physics:.3f} m/s "
          f"({Vs_full_physics / KNOT_TO_MS:.3f} knots)")
```
